#Обнаружение пальмы
Основная часть работы по извлечению функций из изображений состоит в подготовке данных, создании обучающих образцов и обучении модели. Теперь, когда эти шаги выполнены, вы будете использовать обученную модель для обнаружения пальм по всему изображению. Обнаружение объекта - это процесс, который обычно требует нескольких тестов для достижения наилучших результатов. Есть несколько параметров, которые вы можете изменить, чтобы ваша модель работала наилучшим образом. Чтобы быстро проверить эти параметры, вы попытаетесь обнаружить деревья в небольшом разделе изображения. Как только вы будете удовлетворены результатами, вы расширите инструменты обнаружения до полного изображения.

На панели Геообработка найдите и откройте инструмент « Обнаружение объектов с использованием глубокого обучения» .
Этот инструмент вызывает сторонний Python API для глубокого изучения и использует указанную растровую функцию Python для обработки изображения.

Для инструмента « Обнаружение объектов с использованием глубокого обучения» введите следующие параметры:
В качестве входного растра выберите Kolovai Palms .
Для вывода обнаруженных объектов введите CoconutTrees .
Для определения модели перейдите к PalmDetection_25_SSD.emd (загружается с данными урока, расположенными в C: \ DeepLearning \ Data ) или к файлу .emd, который был создан с помощью инструмента Train Deep Learning Model.
Заполнение : 0
Порог : 0,5
nms_overlap : 0,6
В качестве размера партии введите 1.
Установите флажок для Не Максимальное подавление .
Для максимального коэффициента перекрытия введите 0,6 .
Дополнительные аргументы появятся в окне инструмента из-за информации в обученной модели. Специалист по данным, предоставивший вам модель, должен предоставить предложения для каждого аргумента. Дополнительная информация об аргументах предоставляется, если вы хотите поэкспериментировать с другими значениями. В противном случае сохраните настройки по умолчанию.

Порог аргумент является уверенность пороговых сколько доверия приемлемо для обозначения объекта пальмы? Это число можно настроить для достижения желаемой точности.

Выполняя свертку изображений в моделировании сверточных нейронных сетей, вы существенно сжимаете данные, и пиксели на краю изображения используются намного меньше во время анализа по сравнению с внутренними пикселями. По умолчанию параметр заполнения равен 0, но параметр заполнения 1 означает, что к внешним краям изображения добавляется дополнительная граница пикселей со значением 0. Это уменьшает потерю информации из допустимых краевых пикселей и сокращается. Вы можете изменить параметр на 1 или 2, чтобы увидеть эффекты.

Меньшее число для этого аргумента будет указывать, что объекты не могут перекрываться, чтобы считаться отдельными элементами. Размер партииПараметр определяет количество выборок, которые будут использоваться для обучения сети на каждой итерации обучения. Например, если у вас есть 1000 обучающих образцов (чипов изображений) и размер пакета 100, первые 100 обучающих образцов будут обучать нейронную сеть. На следующей итерации будут использоваться следующие 100 сэмплов и так далее. В зависимости от объема памяти, имеющегося на вашей машине, вы можете увеличить этот параметр, хотя обучение следует проводить только в идеальных квадратных партиях. Например, вы можете использовать пакет 4, 9, 16 и далее.

Инструменту может потребоваться некоторое время для запуска полного изображения, поэтому, если вы хотите поэкспериментировать с другими параметрами, вы измените степень обработки на меньшую область.

Увеличьте масштаб изображения до масштаба 1: 500.
Перейдите на вкладку « Среды » в инструменте « Обнаружение объектов с использованием глубокого обучения» . Измените экстент на текущий экстент дисплея .
Нажмите Run .
Попробуйте изменить параметры, чтобы увидеть, улучшается ли ваша модель. Как только вы будете удовлетворены результатами в меньшем масштабе, измените степень обработки обратно на По умолчанию, чтобы обработать все изображение.
Нажмите Run .
Работа инструмента займет несколько минут, в зависимости от вашего оборудования и от того, используете ли вы процессор, графический процессор или оперативную память.

#Оценка  здоровья растительности


Чтобы оценить состояние растительности, вы рассчитаете Индекс видимой атмосферной устойчивости (VARI), который был разработан как косвенная мера индекса площади листа (LAI) и доли растительности (VF), используя только значения отражательной способности от видимой длины волны:

(Rg - Rr) / (Rg + Rr - R(Rg - Rb))

где R r , R g и R b - значения коэффициента отражения для красной, зеленой и синей полос соответственно (Gitelson et al. 2002).

Как правило, для оценки здоровья растительности вы должны использовать значения отражательной способности как в видимом, так и в ближнем инфракрасном (NIR) диапазонах длин волн, как и в случае нормализованного разностного индекса растительности (NDVI). Растровые данные Kolovai Palms, загруженные вами из OpenAerialMap, представляют собой многоканальное изображение с тремя полосами, все в видимом электромагнитном спектре, поэтому вы будете использовать вместо этого VARI.

##Рассчитать VARI


Измерение VARI требует ввода трех полос в растре Kolovai Palms. Чтобы рассчитать VARI, вы будете использовать растровую функцию Band Arithmetic. Растровые функции работают быстрее, чем инструменты геообработки, поскольку они не создают новый набор растровых данных. Вместо этого они выполняют анализ пикселей в реальном времени при панорамировании и масштабировании.

При необходимости откройте свой проект CoconutHealth в ArcGIS Pro .
На ленте щелкните вкладку « Изображения ». В группе Анализ щелкните Растровые функции .
На панели « Растровые функции» найдите и выберите растровую функцию Band Arithmetic .

В функции Band Arithmetic Properties установите следующие параметры и нажмите `Create new layer` :
Для Растра выберите растровый слой `Kolovai Palms` .
Для метода выберите `VARI` . Функция требует, чтобы вы указали индекс полос, который соответствует входным полосам для формулы. Параметр «Вход» под параметром «Индексы диапазонов» отображает «Красный, зеленый, синий», поэтому номера индексов диапазонов, соответствующие полосам «Красный», «Зеленый» и «Синий», будут указаны в указанном порядке. Убедитесь, что между каждой полосой находится один пробел.
Для индексов полос введите `1 2 3` .

На панели Содержание убедитесь, что выбран слой Band Arithmetic_Kolovai Palms .
На ленте щелкните контекстную вкладку « Внешний вид ».

В группе « Рендеринг » выберите раскрывающееся меню « Тип растяжки» и выберите « Стандартное отклонение» .

На панели Содержание дважды щелкните по элементу Arithmetic_Kolovai Palms и переименуйте его в VARI 



##Извлечение VARI  пальм

Наличие растрового слоя, показывающего VARI, полезно, но не обязательно может быть выполнено. Чтобы выяснить, какие деревья нуждаются во внимании, вы хотите знать среднее значение VARI для каждого отдельного дерева. Чтобы найти значение VARI для каждого дерева, вы извлечете базовый средний VARI и обозначите их символами, чтобы показать, какие деревья здоровы, а какие нуждаются в обслуживании.

Сначала вы конвертируете объекты многоугольника в круги для представления пальм.

На панели « Геообработка» найдите и откройте инструмент « Привязать к точке» . Введите следующие параметры и нажмите « Выполнить» :
Для Входных объектов выберите слой `CoconutTrees` .
В качестве выходного класса объектов введите `CoconutTrees_Points` .

У вас есть класс точечных объектов в центре тяжести каждого обнаруженного многоугольника. Если вы увеличите изображение в разных местах и ​​используете инструмент « Измерить» , вы увидите, что средний радиус пальм составляет около 3 метров.

На панели Геообработка найдите и откройте инструмент `Буфер` .
Укажите следующие параметры и нажмите « Выполнить» :
Для Входных объектов выберите CoconutTrees_Points .
Для Выходного класса объектов выберите PalmTreesBuffer .
В качестве расстояния выберите « 3 метра» (убедитесь, что установлено значение « Линейная единица измерения» ).
У вас есть класс объектов полигонов, отображающий расположение и общую форму каждой вершины кокосовой пальмы.

На панели Содержание отключите слои VARI , CoconutTrees и CoconutTrees_Points .

Вы присоедините эту таблицу к слою PalmTreesBuffer, чтобы получить класс пространственных объектов с показателем достоверности и средним значением VARI для каждой обнаруженной пальмы.

Для входной таблицы выберите PalmTreesBuffer .
Для поля ввода соединения выберите ORIG_FID .
Для таблицы соединений выберите MeanVARI_per_Palm .
В поле «Выходное соединение» выберите « ORIG_FID» .
Для полей объединения выберите MEAN .

К слою PalmTreesBuffer теперь добавлено поле MEAN . Вы переименуете его и символизируете для лучшего понимания данных.

В Contents панели нажмите PalmTreesBuffer два раза и переименовать его PalmTreesVARI .

Для поля выберите MEAN .
